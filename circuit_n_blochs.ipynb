```python
# the circuit used and the bloch sphere representation of the qubit stateds
!pip install qiskit qiskit-aer matplotlib
!pip install pylatexenc
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector
import math
import matplotlib.pyplot as plt

#angles used for actual sim r as belows
low_a = math.pi / 3
high_a = 2 * math.pi / 3

#since there r 2 angles...the 2 circuits corresponding to them would be
qc_low = QuantumCircuit(1)
qc_low.ry(low_a, 0)


qc_high = QuantumCircuit(1)
qc_high.ry(high_a, 0)
```

```python
from IPython.display import display

print("Qc for ry(pi/3):")
fig1 = qc_low.draw('mpl')
display(fig1)

print("Qc for ry(2*pi/3):")
fig2 = qc_high.draw('mpl')
display(fig2)
```

```python
#bloch sphere reprsenting the state vector of qubit
state_low = Statevector(qc_low)
print("Bloch sphere for ry(pi/3):")
fig3 = plot_bloch_multivector(state_low, title="Bloch Sphere for ry(pi/3)")
display(fig3)
```

```python
state_high = Statevector(qc_high)
print("Bloch sphere for ry(2*pi/3):")
fig4 = plot_bloch_multivector(state_high, title="Bloch Sphere for ry(2*pi/3)")
display(fig4)
```
